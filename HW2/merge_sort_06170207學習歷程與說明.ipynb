###                   mergesort若將焦點放在合併，就是將兩個已排序好的數列合併成一個，然而今天要做的是將一個未排序過的數列變成一個整理好的數列，在找到下面那個參考網站後，我對於mergesort有了初步的想法，因只有一個數列，所以首先要做的是分割，分割完之後再做合併，所以我打算將其分兩部分來寫，第一個是分割的程式，如下:

```python
def MergeSort(self, arr):
            
        length = len(arr)
        if length > 1:
            mid = len(arr) // 2 #取整數
            L = arr[0:mid] #取得左半邊的資料
            R = arr[mid:len(arr)] #取得右半邊的資料
            self.MergeSort(L) #將左右兩邊丟回繼續分割
            self.MergeSort(R)
            return Merge(arr,L,R) #丟到Merge做比較與合併
```

### 因要把數列分割到每個都只剩下一個數字，所以利用長度大於1的條件去讓他跑，而分割的做法就是不斷得分兩半，設mid 為一半中間的那個數字，再設定陣列 L 與 R 分別取得左右兩邊的陣列，最後再重新呼叫丟回繼續分割。

### 再來是合併並且排序，程式碼如下:

```python
def Merge(arr,L,R):
        
        a = arr
        lindex = rindex = aindex = 0 #取得左右及原陣列的索引
        while lindex < len(L) and rindex < len(R): 
            if L[lindex] < R[rindex]: #左右開始比較
                a[aindex] = L[lindex] #左邊較小則把左邊的值賦予回原陣列做替換
                lindex += 1 #左邊索引往右走
            else:
                a[aindex] = R[rindex] #反之就將右邊的值賦予回原陣列
                rindex += 1
            aindex += 1 #原陣列往右
        while lindex < len(L):
            a[aindex] = L[lindex]
            lindex += 1
            aindex += 1
        while rindex < len(R):
            a[aindex] = R[rindex]
            rindex += 1
            aindex += 1
        return arr
    
               
    
        
```
